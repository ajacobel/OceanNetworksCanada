# Spectrogram: Visualizing Sound

A spectrogram shows audio data as spectral intensities over time. On the y axis you can see frequencies, which are plotted over time on the x axis. The colours indicate power or loudness of the sound signals recorded. In the spectrogram below, the yellow and bright green time periods had the loudest sounds
